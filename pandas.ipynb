**Задание 10. Ответьте на вопросы:**
    * Какая доля студентов сдала экзамен по математике (passmark >= 50)?
    * Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?
    * Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?

```python
# Доля студентов, сдавших экзамен по математике (passmark >= 50)
passed_frac = len(df[df['math score'] >= passmark]) / len(df)

# Доля студентов, проходивших курс подготовки к экзамену и сдавших экзамен по математике
prep_passed_frac = len(df[(df['test preparation'] == 'completed') & (df['math score'] >= passmark)]) / len(df[df['test preparation'] == 'completed'])

# Доля женщин, не проходивших курс подготовки и не сдавших экзамен по математике
failed_female_frac = len(df[(df['test preparation'] == 'none') & (df['gender'] == 'female') & (df['math score'] < passmark)]) / len(df[(df['test preparation'] == 'none') & (df['gender'] == 'female')])

# Вывод результатов
print("Доля сдавших по математике:", passed_frac)
print("Доля сдавших по математике с курсом подготовки:", prep_passed_frac)
print("Доля женщин, не сдавших по математике без подготовки:", failed_female_frac)
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
import time

# Задание 12.1: Средний балл за экзамен по чтению для каждой этнической группы с использованием циклов
start_time = time.time()
ethnic_reading_mean = {}
ethnic_groups = df['race/ethnicity'].unique()
for ethnic_group in ethnic_groups:
    total_score = 0
    count = 0
    for index, row in df.iterrows():
        if row['race/ethnicity'] == ethnic_group:
            total_score += row['reading score']
            count += 1
    ethnic_reading_mean[ethnic_group] = total_score / count if count != 0 else 0
end_time = time.time()

print("Средний балл за экзамен по чтению для каждой этнической группы:")
print(ethnic_reading_mean)
print("Время выполнения задания:", end_time - start_time, "секунд")

# Задание 12.2: Минимальный балл за экзамен по письму для каждого уровня образования с использованием циклов
start_time = time.time()
education_writing_min = {}
education_levels = df['education'].unique()
for education_level in education_levels:
    min_score = float('inf')
    for index, row in df.iterrows():
        if row['education'] == education_level:
            min_score = min(min_score, row['writing score'])
    education_writing_min[education_level] = min_score
end_time = time.time()

print("\nМинимальный балл за экзамен по письму для каждого уровня образования:")
print(education_writing_min)
print("Время выполнения задания:", end_time - start_time, "секунд")
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
# Создание столбца Math_PassStatus
df['Math_PassStatus'] = df['math score'].apply(lambda x: 'P' if x >= passmark else 'F')

# Создание столбца Reading_PassStatus
df['Reading_PassStatus'] = df['reading score'].apply(lambda x: 'P' if x >= passmark else 'F')

# Создание столбца Writing_PassStatus
df['Writing_PassStatus'] = df['writing score'].apply(lambda x: 'P' if x >= passmark else 'F')

# Подсчет количества студентов, сдавших и не сдавших экзамены
math_pass_count = df['Math_PassStatus'].value_counts()
reading_pass_count = df['Reading_PassStatus'].value_counts()
writing_pass_count = df['Writing_PassStatus'].value_counts()

print("Количество студентов, сдавших и не сдавших экзамен по математике:")
print(math_pass_count)
print("\nКоличество студентов, сдавших и не сдавших экзамен по чтению:")
print(reading_pass_count)
print("\nКоличество студентов, сдавших и не сдавших экзамен по письму:")
print(writing_pass_count)
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark > 90:
        return 'A'
    elif 80 <= average_mark <= 90:
        return 'B'
    elif 70 <= average_mark < 80:
        return 'C'
    elif 60 <= average_mark < 70:
        return 'D'
    elif 50 <= average_mark < 60:
        return 'E'
    else:
        return 'F'

# Создание столбца Grade с оценками для каждого студента
df['Grade'] = df[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)

# Вывод количества студентов, получивших каждую из оценок
grade_counts = df['Grade'].value_counts()

print("Количество студентов, получивших каждую из оценок:")
print(grade_counts)
```
